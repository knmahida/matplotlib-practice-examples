```python
# Import matplotlib.pyplot as plt for plotting

import matplotlib.pyplot as plt
```

```python
# Enable inline plotting in Jupyter notebooks

%matplotlib inline
```

```python
# Create sample data x and y where y = x**2

import numpy as np
x = np.linspace(0, 10, 100)
y = x**2
```

```python
# Display the x array (inspect values)

x
```

```python
# Display the y array (inspect values)

y
```

```python
# Basic plot of y vs x with labels and title

plt.plot(x, y)
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
```

```python
# Show two subplots side-by-side with different line styles

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-')
plt.title("Red Line")
plt.subplot(1, 2, 2)
plt.plot(x, y, 'b--')
plt.title("Blue Dashed Line")
```

```python
# Create a figure and single axes using add_axes, then plot data and label axes

fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
ax.set_title("Title")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
```

```python
# Create a figure with a larger main axes and a smaller inset axes, each with their own plot

fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(x,y)
ax1.set_title("Larger Plot")
ax2.plot(y,x)
ax2.set_title("Smaller Plot")
```

```python
# Create a 1x2 set of subplots and return the figure and axes array

fig,axs3=plt.subplots(nrows=1,ncols=2)
```

```python
# Display the axes array object to inspect its structure

axs3
```

<!-- Section header introducing figure size and DPI examples -->

Figure Size and DPI

```python
# Create a 2x3 grid of subplots with specified figure size and DPI, then adjust layout

fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(10,7),dpi=100)
axes
plt.tight_layout()
```

```python
# Save the current figure to disk as 'my_figure.png' with specified DPI

fig.savefig("my_figure.png",dpi=200)
```

```python
# Plot x squared and x cubed on the same axes and show a legend

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x squared")
ax.plot(x,x**3,label="x cubed")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Title")
ax.legend()  # To show the legend
```
